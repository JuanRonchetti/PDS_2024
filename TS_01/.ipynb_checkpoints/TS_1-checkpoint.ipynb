## Enunciado
La tarea consiste en programar una función que genere señales senoidales y que permita parametrizar:

* Amplitud máxima de la senoidal (volts)
* Valor medio (volts)
* Frecuencia (Hz)
* Fase (radianes)
* Cantidad de muestras digitalizada por el ADC (muestras)
* Frecuencia de muestreo del ADC (Hz)

La función debe admitir ser llamada de la siguiente manera:
* tt, xx = mi_funcion_sen( vmax = 1, dc = 0, ff = 1, ph=0, nn = N, fs = fs)

Bonus:
* Implementar alguna otra señal propia de un generador de señales. 

## Resolucion

```python
# -------------------------------------------------------------------------------
#                                   Imports
# -------------------------------------------------------------------------------
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import scipy.signal as sig
import matplotlib as mpl
import matplotlib.pyplot as plt

# -------------------------------------------------------------------------------
#                                   Funcion
# -------------------------------------------------------------------------------

def my_signal_generator( ff, nn, fs, vmax = 1, dc = 0, ph = 0, signal = 'sine' ):
    
    # Tiempo de muestreo
    ts = 1/fs
    
    # Sampleo temporal
    tt = np.linspace(0, (nn-1)*ts, nn).flatten()
    
    # Armado de senial
    if signal == 'square':
        
        aux = vmax * sig.square(2 * np.pi * ff * tt + ph) + dc
        xx = aux.reshape(nn,1)
        return (tt, xx)
        
    elif signal == 'sawtooth':
        
        aux = vmax * sig.sawtooth(2 * np.pi * ff * tt + ph) + dc
        xx = aux.reshape(nn,1)
        return (tt, xx)
    
    # Si no se especifica nada (o algo erroneo) se devuelve un seno
    else:
        
        aux = vmax * np.sin( 2*np.pi*ff*tt + ph ) + dc
        xx = aux.reshape(nn,1)
        return (tt, xx)
```

## Implementacion

```python
# Invoca la funcion
tt, xx = my_signal_generator( ff = 1, nn = 1000, fs = 1000 )

# Presentación gráfica de los resultados
plt.plot(tt, xx)
plt.title('Señal senoidal con parametros por defecto')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud [V]')
plt.show()
```

```python
# Invoca la funcion
tt, xx = my_signal_generator( vmax = 1, dc = 4, ff = 4, ph=np.pi/2, nn = 1000, fs = 1000 )

# Presentación gráfica de los resultados
plt.plot(tt, xx)
plt.title('Señal senoidal con parametros especificados')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud [V]')
plt.show()
```

```python
# Invoca la funcion
tt, xx = my_signal_generator( vmax = 1, dc = 0, ff = 3, ph=0, nn = 1000, fs = 1000, signal='square' )

# Presentación gráfica de los resultados
plt.plot(tt, xx)
plt.title('Señal cuadrada')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud [V]')
plt.show()
```

```python
# Invoca la funcion
tt, xx = my_signal_generator( vmax = 10, dc = 20, ff = 4, ph=0, nn = 1000, fs = 1000, signal='sawtooth' )

# Presentación gráfica de los resultados
plt.plot(tt, xx)
plt.title('Señal diente de sierra')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud [V]')
plt.show()
```
