# 1. Introduccion

Se necesita eliminar las interferencias de una señal electrocardiográfica (ECG) que fue registrada a una frecuencia de muestreo de 1 kHz. 

Para ello, en el trabajo anterior se propuso una plantilla de diseño y se diseñaron filtros IIR y FIR lineales acordes a esta, los cuales fueron utilizados para filtrar la señal. En este trabajo, el objetivo es realizar un filtrado con tecnicas no lineales, y analizar las diferencias y similitudes.

Se pide utilizar las siguientes metodologías no lineales de estimación-substracción para la remoción de la interferencia de línea de base, y comparar con las técnicas lineales del trabajo anterior en las mismas regiones. Luego discutir las ventajas y desventajas de cada técnica.

### Parte A: Estimación del movimiento de línea de base a partir del filtrado de mediana 
A partir de la siguiente estimación del movimiento de línea de base:

$$ \hat{b}={\text{med}_{600}\{\text{med}_{200}\{s\}\}} $$
Donde $s$ representa la señal de ECG a partir de la que estimamos el movimiento de la línea de base $\hat{b}$. 
El operador med200 calcula muestra a muestra la mediana en los 200 ms previos a la muestra actual. Luego dicha estimación la substraemos para estimar la actividad eléctrica cardíaca $\hat{x}$ tal que:
$$\hat{x} = s - \hat{b}$$

### Parte B: Estimación del movimiento de línea de base a partir de los segmentos isoeléctricos de cada latido 
A partir de las detecciones provistas, se puede estimar aproximadamente el nivel isoeléctrico, es decir el momento donde la actividad eléctrica cardíaca es nula. El intervalo entre la onda P y Q, es un buen momento para estimar las características de las señales interferentes que originan el movimiento de la línea de base.
Para la localización del segmento PQ es posible valerse de la localización de cada latido y a partir de la misma, anticipar un tiempo adecuado de manera tal de no incluir la onda P ni Q. 
La señal de línea de base debe estar muestreada a la misma frecuencia que el ECG, por lo que deberá remuestrear dicha señal.

# 2. Inicializacion de bibliotecas y funciones

```python
# -------------------------------------------------------------------------------
#                                Imports
# -------------------------------------------------------------------------------
import sympy as sp
import numpy as np
import scipy.signal as sig
import scipy.io as sio
import matplotlib.pyplot as plt
from pytc2.sistemas_lineales import plot_plantilla
from scipy.interpolate import CubicSpline

import warnings
warnings.filterwarnings('ignore')

# -------------------------------------------------------------------------------
#                                Datos de ECG
# -------------------------------------------------------------------------------
mat_struct = sio.loadmat('ecg.mat')
ecg_one_lead = mat_struct['ecg_lead'].flatten()
cant_muestras = len(ecg_one_lead)

qrs_detections = mat_struct['qrs_detections'].flatten()

fs=1000
nyq = fs/2

fig_sz_x = 8
fig_sz_y = 3
fig_dpi = 100 # dpi

regs_interes = ( 
        np.array([5, 5.2]) *60*fs, # minutos a muestras
        np.array([12, 12.4]) *60*fs, # minutos a muestras
        np.array([15, 15.2]) *60*fs, # minutos a muestras
        [4000, 5500], # muestras
        [10e3, 11e3], # muestras
)

# -------------------------------------------------------------------------------
#                       Funcion para graficar post-filtrado
# -------------------------------------------------------------------------------
def plot_ecg_filtered(filtered_signal, title, demora=0,
                      fig_sz_x=8, fig_sz_y=3, fig_dpi=100):
    # Crear la figura y los ejes para los subplots
    num_regs = len(regs_interes)
    fig, axs = plt.subplots(num_regs, 1, figsize=(fig_sz_x, fig_sz_y * num_regs), dpi=fig_dpi)

    for ii, ax in zip(regs_interes, axs):
        zoom_region = np.arange(np.max([0, ii[0]]), np.min([len(ecg_one_lead), ii[1]]), dtype='uint')

        # Graficar señal ECG original
        ax.plot(zoom_region, ecg_one_lead[zoom_region], label='ECG', linewidth=2)

        # Graficar señal filtrada con la demora aplicada
        ax.plot(zoom_region, filtered_signal[zoom_region + demora], label=title)

        # Configuración del gráfico
        ax.set_title(f'{title} de {ii[0]} a {ii[1]}')
        ax.set_ylabel('Adimensional')
        ax.set_xlabel('Muestras (#)')
        
        ax.legend()
        ax.set_yticks(())

    plt.tight_layout()  # Ajustar el espaciado entre subplots
    plt.show()
```

# 3. Filtros Lineales del Trabajo Anterior

## 3.1 Filtro IIR

```python
# -------------------------------------------------------------------------------
#                                IIR
# -------------------------------------------------------------------------------
fs = 1000
nyq = fs/2

#------ Plantilla ------
wp = np.array([0.5, 30]) / nyq
ws = np.array([0.1, 45]) / nyq
gpass = 0.05
gstop = 40

num, den = sig.iirdesign(wp, ws, gpass, gstop, analog=False, ftype='cheby1', output='ba')
sos_iir = sig.iirdesign(wp, ws, gpass, gstop, analog=False, ftype='cheby1', output='SOS')

order = len(num) - 1  # Orden del filtro

wrad, hh = sig.sosfreqz(sos_iir, worN=10000)
ww = wrad / np.pi * nyq

gd = -np.diff(np.unwrap(np.angle(hh))) / np.diff(ww)  # Retardo en muestras
wwgd = ww[1:]  # Ajustar ww para que coincida con gd

t = np.arange(0, 10, 0.001)
imp = np.fft.ifft(hh)

#------------------------------- Graficos -------------------------------

#------ Modulo ------
modulo = plt.figure(figsize=(12, 4))
plt.title(f'Filtro IIR orden {order}: Modulo')
plt.plot(ww, 20 * np.log10(abs(hh)))
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Modulo [dB]')
plt.grid()
plt.axis([0, 100, -50, 5])
plot_plantilla(filter_type='bandpass', fpass=[wp[0] * nyq, wp[1] * nyq], ripple=gpass,
               fstop=[ws[0] * nyq, ws[1] * nyq], attenuation=gstop, fs=fs)
plt.show()

#------ Filtrado ------
#ECG_f_iir = sig.sosfilt(sos_iir, ecg_one_lead)
#plot_ecg_filtered(ECG_f_iir, 'Filtrado IIR',fig_sz_x=8, fig_sz_y=3)

ECG_f_iir_b = sig.sosfiltfilt(sos_iir, ecg_one_lead)
plot_ecg_filtered(ECG_f_iir_b, 'Filtrado IIR Bidireccional', fig_sz_x=12, fig_sz_y=4)
```

## 3.2 Filtro FIR

```python
# -------------------------------------------------------------------------------
#                                FIR
# -------------------------------------------------------------------------------
fs = 1000
nyq = fs/2

frecs = [0.0,     ws[0]*nyq,    wp[0]*nyq,       wp[1]*nyq,    ws[1]*nyq,       nyq]
gains = [-gstop,   -gstop,       -gpass,         -gpass,      -gstop,      -gstop] 
gains = 10**(np.array(gains)/20)

den_f = 1.0
cant_coef = 3001

fir_win = sig.firwin2(cant_coef, frecs, gains , window='blackmanharris', fs=fs)

wrad, hh = sig.freqz(fir_win, den_f, worN=10000)
ww = wrad / np.pi * nyq

gd = -np.diff(np.unwrap(np.angle(hh))) / np.diff(ww)  # Retardo en muestras
ww_gd = ww[1:]  # Ajustar ww para que coincida con gd

t = np.arange(0, 10, 0.001)
imp = np.fft.ifft(fir_win)  # Usa los coeficientes del filtro FIR directamente

#------------------------------- Graficos -------------------------------

#------ Modulo ------
modulo = plt.figure(figsize=(12, 4))
plt.title(f'Filtro FIR orden {cant_coef}: Modulo')
plt.plot(ww, 20 * np.log10(abs(hh)))
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Modulo [dB]')
plt.axis([0, 100, -50, 5])
plt.grid()
plot_plantilla(filter_type='bandpass', fpass=[frecs[2], frecs[3]], ripple=gpass,
               fstop=[frecs[1], frecs[4]], attenuation=gstop, fs=fs)
plt.show()

#------ Filtrado ------
ECG_f_win = sig.lfilter(fir_win, den_f, ecg_one_lead)
demora_fir = int((cant_coef - 1) / 2)
plot_ecg_filtered(ECG_f_win, 'Filtrado FIR', demora = demora_fir, fig_sz_x=12, fig_sz_y=4)
```

# 5. Estimacion del Nivel Isoeléctrico

```python
# -------------------------------------------------------------------------------
#               Referencia para la determinacion del inicio del latido
# -------------------------------------------------------------------------------
def plot_ecg_with_qrs(ecg_one_lead, qrs_detections, reg_interes, fs=1000):
    # Definir la región de zoom
    zoom_region = np.arange(np.max([0, reg_interes[0]]), np.min([len(ecg_one_lead), reg_interes[1]]), dtype='uint')

    # Graficar resultados
    plt.figure(figsize=(8, 5))
    plt.plot(zoom_region, ecg_one_lead[zoom_region], label='Señal ECG', linewidth=2)

    # Graficar líneas verticales para las detecciones QRS alineadas con la región de zoom
    for detection in qrs_detections:
        if reg_interes[0] <= detection < reg_interes[1]:  # Solo graficar si está dentro de la región
            plt.axvline(x=detection, color='red', linestyle='--', label='Detección QRS' if detection == qrs_detections[0] else "")

            x2 = max(0, detection - int(0.05 * fs))  # 50 ms antes del QRS
            plt.axvline(x=x2, color='blue', linestyle=':', label='Inicio Latido' if x2 == detection - int(0.05 * fs) else "")
            
    plt.title('Referencia para Interpolacion')
    plt.xlabel('Muestras (#)')
    plt.ylabel('Amplitud')
    plt.legend()
    plt.grid()
    plt.show()
    return x2

# Grafico
reg_interes=[10e3, 10.4e3]
x2 = plot_ecg_with_qrs(ecg_one_lead, qrs_detections, reg_interes)
```

```python
# -------------------------------------------------------------------------------
#                      Filtrado con interpolacion cubica
# -------------------------------------------------------------------------------
def cubicspline_filter(ecg_data, qrs_d):
    # Filtrar puntos válidos para la interpolación
    valid_points = [(int(r - 50), ecg_data[int(r - 50)]) for r in qrs_d if r >= 50]
    
    # Descomponer los puntos en x e y
    x_points, y_points = zip(*valid_points) if valid_points else ([], [])
    
    # Asegurar que x_points y y_points estén ordenados
    sorted_indices = np.argsort(x_points)
    x_points = np.array(x_points)[sorted_indices]
    y_points = np.array(y_points)[sorted_indices]

    # Crear el spline cúbico
    spline = CubicSpline(x_points, y_points)
    
    # Filtrar la señal original
    filtered = ecg_data - spline(np.arange(len(ecg_data)))

    return filtered, (x_points, y_points), spline

# Filtrar la señal ECG
ecg_one_lead_filtered, points, spline = cubicspline_filter(ecg_one_lead, qrs_detections)

#------------------------------- Graficos -------------------------------
for ii in regs_interes:
    zoom_region = np.arange(np.max([0, ii[0]]), np.min([cant_muestras, ii[1]]), dtype='uint')
    
    plt.figure(figsize=(12, 4), dpi= fig_dpi, facecolor='w', edgecolor='k')
    plt.plot(zoom_region, ecg_one_lead[zoom_region], label='ECG', linewidth=2)
    plt.plot(zoom_region, ecg_one_lead_filtered[zoom_region], label='Filtrado por Interpolacion', linewidth=2)
    plt.plot(zoom_region, spline(np.arange(len(ecg_one_lead)))[zoom_region], label='Spline')
    
    valid_indices = (points[0] >= ii[0]) & (points[0] <= ii[1])
    plt.plot(points[0][valid_indices], points[1][valid_indices], "rx", label="Puntos de QRS")
    
    plt.title('ECG desde ' + str(ii[0]) + ' a ' + str(ii[1]) )
    plt.ylabel('Adimensional')
    plt.xlabel('Muestras (#)')
    
    axes_hdl = plt.gca()
    axes_hdl.legend()
    axes_hdl.set_yticks(())
            
    plt.show()
```

# 7. Analisis de resultados
A partir de los graficos obtenidos, se pueden señalar las siguientes caracteristicas:
#### Ventajas de los filtros no lineales:
* Robustez ante Ruido: los filtros no lineales menos sensible a los picos y valores atípicos en comparación con métodos lineales. Se obtiene una mejor estimacion de la linea de base.
* Preservación de Picos: Mantienen las características importantes del ECG, como los picos R, sin distorsionarlos como en el caso de los filtros lineales.
#### Desventajas de los filtros no lineales:
* Computacionalmente Costosos: Requieren más tiempo computacional debido a las operaciones sobre ventanas múltiples. Los filtros lineales son mas faciles de implementar y computacionalmente mas eficienetes.
* Demora: los filtros no lineales requieren acceso a la señal completa, por ende no pueden ser implementados en tiempo real, a diferencia de los filtros lineales.
