# Introduccion

Se necesita eliminar las interferencias de una señal electrocardiográfica (ECG) que fue registrada a una frecuencia de muestreo de 1 kHz. Para ello se propone la siguiente plantilla de diseño:

* ws1 = 0.1 Hz
* wp1 = 0.5 Hz
* wp2 = 30.0 Hz
* ws2 = 45.0 Hz

### Parte 1: Diseño

Se pide diseñar dos filtros: uno IIR y otro FIR que cumplan con la plantilla. Mostrar la respuesta en frecuencia, para ambos filtros, de:
* Módulo
* Fase
* Retardo
* Respuesta al impulso

### Parte 2: Implementacion

En esta parte se utilizaran los dos filtros diseñados para mitigar las interferencias de la señal dada. Esta corresponde a una medicion de electrocardiograma de superficie (ECG). 

## Inicializacion de bibliotecas

```python
# -------------------------------------------------------------------------------
#                                Imports
# -------------------------------------------------------------------------------
import sympy as sp
import numpy as np
import scipy.signal as sig
import scipy.io as sio
import matplotlib.pyplot as plt
from pytc2.sistemas_lineales import  analyze_sys, plot_plantilla

import warnings
warnings.filterwarnings('ignore')
```

## Parámetros y Plantilla

```python
# -------------------------------------------------------------------------------
#                                Inicializacion
# -------------------------------------------------------------------------------
fs = 1000
nyq = fs/2

#------ Plantilla ------
wp = np.array([0.5, 30]) / nyq
ws = np.array([0.1, 45]) / nyq
gpass = 0.05
gstop = 40

frecs = [0.0,     ws[0]*nyq,    wp[0]*nyq,       wp[1]*nyq,    ws[1]*nyq,       nyq]
gains = [-gstop,   -gstop,       -gpass,         -gpass,      -gstop,      -gstop] 
gains = 10**(np.array(gains)/20)

plt.title('Plantilla')
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Modulo [dB]')
plt.grid()
plt.axis([0,100,-50, 5])
plot_plantilla(filter_type = 'bandpass', fpass = [wp[0]* nyq, wp[1]* nyq ], ripple = gpass , 
               fstop = [ws[0]* nyq, ws[1]* nyq ], attenuation = gstop, fs = fs)
```

# Parte 1 
## Diseño de Filtro IIR
En esta seccion se diseñara un filtro IIR que cumpla con la plantilla solicitada, haciendo uso de la funcion iirdesign de la biblioteca scipy signal de python:

```python
# -------------------------------------------------------------------------------
#                                Disenio IIR
# -------------------------------------------------------------------------------

num, den = sig.iirdesign(wp, ws, gpass, gstop, analog=False, ftype='cheby1', output='ba')
sos_iir = sig.iirdesign(wp, ws, gpass, gstop, analog=False, ftype='cheby1', output='SOS')

order = len(num) - 1  # Orden del filtro

wrad, hh = sig.sosfreqz(sos_iir, worN=10000)
ww = wrad / np.pi * nyq

gd = -np.diff(np.unwrap(np.angle(hh))) / np.diff(ww)  # Retardo en muestras
wwgd = ww[1:]  # Ajustar ww para que coincida con gd

t = np.arange(0, 10, 0.001)
imp = np.fft.ifft(hh)

#------------------------------- Graficos -------------------------------

#------ Modulo ------
modulo = plt.figure(figsize=(10, 6))
plt.title(f'Filtro IIR orden {order}: Modulo')
plt.plot(ww, 20 * np.log10(abs(hh)))
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Modulo [dB]')
plt.grid()
plt.axis([0, 100, -50, 5])
plot_plantilla(filter_type='bandpass', fpass=[wp[0] * nyq, wp[1] * nyq], ripple=gpass,
               fstop=[ws[0] * nyq, ws[1] * nyq], attenuation=gstop, fs=fs)
plt.show()

#------ Modulo (Zoom entre 0 y 1 Hz) ------
plt.figure(figsize=(10, 6))
plt.title(f'Filtro IIR orden {order}: Modulo (Zoom)')
plt.plot(ww, 20 * np.log10(abs(hh)))
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Módulo [dB]')
plt.grid()
plt.axis([0, 1, -50, 5])
plot_plantilla(filter_type='bandpass', fpass=[wp[0] * nyq, wp[1] * nyq], ripple=gpass,
               fstop=[ws[0] * nyq, ws[1] * nyq], attenuation=gstop, fs=fs)
plt.show()

#------ Fase ------
fase = plt.figure(figsize=(10, 6))
plt.title(f'Filtro IIR orden {order}: Fase')
angles = np.unwrap(np.angle(hh))
plt.plot(ww, angles)
plt.ylabel('Fase [rads]')
plt.xlabel('Frecuencia [Hz]')
plt.grid()
plt.axis([0, 100, -50, 5])
plt.show()

#------ Retardo ------
ret = plt.figure(figsize=(10, 6))
plt.title(f'Filtro IIR orden {order}: Retardo')
plt.plot(wwgd, gd)
plt.ylabel('Retardo [s]')
plt.xlabel('Frecuencia [Hz]')
plt.axis([0, 100, 0, 40])
plt.grid()
plt.show()

#------ Respuesta al impulso ------
impulso = plt.figure(figsize=(10, 6))
plt.title(f'Filtro IIR orden {order}: Respuesta al Impulso')
plt.plot(t, imp)
plt.ylabel('Amplitud')
plt.xlabel('Tiempo [s]')
plt.xlim(0, 1)
plt.grid()
plt.show()
```

## Diseño de Filtro FIR
En esta seccion se diseñara un filtro FIR que cumpla con la plantilla solicitada, haciendo uso de la funcion firwin2 de la biblioteca scipy signal de python:

```python
#------ Disenio ------

den_f = 1.0
cant_coef = 3001

fir_win = sig.firwin2(cant_coef, frecs, gains , window='blackmanharris', fs=fs)

wrad, hh = sig.freqz(fir_win, den_f, worN=10000)
ww = wrad / np.pi * nyq

gd = -np.diff(np.unwrap(np.angle(hh))) / np.diff(ww)  # Retardo en muestras
ww_gd = ww[1:]  # Ajustar ww para que coincida con gd

t = np.arange(0, 10, 0.001)
imp = np.fft.ifft(fir_win)  # Usa los coeficientes del filtro FIR directamente

#------------------------------- Graficos -------------------------------

#------ Modulo ------
modulo = plt.figure(figsize=(10, 6))
plt.title(f'Filtro FIR orden {cant_coef}: Modulo')
plt.plot(ww, 20 * np.log10(abs(hh)))
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Modulo [dB]')
plt.axis([0, 100, -50, 5])
plt.grid()
plot_plantilla(filter_type='bandpass', fpass=[frecs[2], frecs[3]], ripple=gpass,
               fstop=[frecs[1], frecs[4]], attenuation=gstop, fs=fs)
plt.show()

#------ Modulo (Zoom entre 0 y 1 Hz) ------
plt.figure(figsize=(10, 6))
plt.title(f'Filtro FIR orden {cant_coef}: Modulo (Zoom)')
plt.plot(ww, 20 * np.log10(abs(hh)))
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Módulo [dB]')
plt.grid()
plt.axis([0, 1, -50, 5])
plot_plantilla(filter_type='bandpass', fpass=[wp[0] * nyq, wp[1] * nyq], ripple=gpass,
               fstop=[ws[0] * nyq, ws[1] * nyq], attenuation=gstop, fs=fs)
plt.show()

#------ Fase ------
fase = plt.figure(figsize=(10, 6))
plt.title(f'Filtro FIR orden {cant_coef}: Fase')
angles = np.unwrap(np.angle(hh))
plt.plot(ww, angles)
plt.ylabel('Fase [rads]')
plt.xlabel('Frecuencia [Hz]')
plt.grid()
plt.show()

#------ Retardo ------
plt.figure(figsize=(10, 6))
plt.title(f'Filtro FIR orden {cant_coef}: Retardo')

valor_promedio_gd = np.mean(gd)
plt.plot(ww_gd, gd, label='Retardo de Grupo')
plt.axhline(y=valor_promedio_gd, color='r', linestyle='--', label='Valor Promedio del Retardo')

# Etiquetas y leyenda
plt.ylabel('Retardo [s]')
plt.xlabel('Frecuencia [Hz]')
plt.grid()
plt.ylim(valor_promedio_gd - 0.01, valor_promedio_gd + 0.01)  # Ajustar límites del eje y alrededor del promedio
plt.legend()
plt.show()

#------ Respuesta al impulso ------
fig, axs = plt.subplots(1, 2, figsize=(10, 6), dpi=100)

#------ Respuesta al Impulso ------
axs[0].plot(np.arange(len(fir_win)), fir_win)
axs[0].set_title(f'Filtro FIR orden {cant_coef}: Respuesta al Impulso')
axs[0].set_ylabel('Amplitud')
axs[0].set_xlabel('Coeficientes del Filtro')
axs[0].grid()

#------ Zoom en la Respuesta al Impulso ------
zoom_region = np.arange(1450, 1551)  # Índices de zoom
axs[1].stem(zoom_region, fir_win[zoom_region], use_line_collection=True)
axs[1].set_title('Zoom en Coeficientes del Filtro')
axs[1].set_ylabel('Amplitud')
axs[1].set_xlabel('Coeficientes del Filtro')
axs[1].grid()

# Ajustar el layout para evitar solapamientos
plt.tight_layout()
plt.show()
```

# Parte 2
Habiendo diseñado exitosamente ambos filtros digitales, se procede a aplicarlos a la señal de ECG dada.
## Levantamiento de datos de ECG y Preparacion de Funcion de Filtrado

```python
# -------------------------------------------------------------------------------
#                                Datos de ECG
# -------------------------------------------------------------------------------
mat_struct = sio.loadmat('ecg.mat')
ecg_one_lead = mat_struct['ecg_lead'].flatten()
cant_muestras = len(ecg_one_lead)

fig_sz_x = 10
fig_sz_y = 7
fig_dpi = 100 # dpi

regs_interes = ( 
        np.array([5, 5.2]) *60*fs, # minutos a muestras
        np.array([12, 12.4]) *60*fs, # minutos a muestras
        np.array([15, 15.2]) *60*fs, # minutos a muestras
        [4000, 5500], # muestras
        [10e3, 11e3], # muestras
)

# -------------------------------------------------------------------------------
#                       Funcion para graficar post-filtrado
# -------------------------------------------------------------------------------

def plot_ecg_filtered(filtered_signal, title, demora=0,
                     fig_sz_x=10, fig_sz_y=5, fig_dpi=100):
    
    for ii in regs_interes:
        zoom_region = np.arange(np.max([0, ii[0]]), np.min([len(ecg_one_lead), ii[1]]), dtype='uint')
        
        plt.figure(figsize=(fig_sz_x, fig_sz_y), dpi=fig_dpi, facecolor='w', edgecolor='k')
        
        # Graficar señal ECG original
        plt.plot(zoom_region, ecg_one_lead[zoom_region], label='ECG', linewidth=2)
        
        # Graficar señal filtrada con la demora aplicada
        plt.plot(zoom_region, filtered_signal[zoom_region + demora], label=title)

        # Configuración del gráfico
        plt.title(f'{title} de {ii[0]} a {ii[1]}')
        plt.ylabel('Adimensional')
        plt.xlabel('Muestras (#)')
        
        axes_hdl = plt.gca()
        axes_hdl.legend()
        axes_hdl.set_yticks(())
                
        plt.show()
```

## Filtrado con filtro IIR

```python
ECG_f_iir = sig.sosfilt(sos_iir, ecg_one_lead)

plot_ecg_filtered(ECG_f_iir, 'Filtrado IIR')
```

### Observaciones
Se observa que si bien el filtro diseñado esta eliminando las bajas frecuencias, esta introduciendo en el sistema una fuerte distorsion. Para analizar si se trata de distorsion de fase introducida durante el filtrado, a continuacion se realizara un filtrado bidireccional de la señal con el mismo filtro, y se compararán los resultados.

## Filtrado Bidireccional con IIR

```python
ECG_f_iir_b = sig.sosfiltfilt(sos_iir, ecg_one_lead)

plot_ecg_filtered(ECG_f_iir_b, 'Filtrado IIR Bidireccional')
```

### Observaciones
Se observa una mucho mejor respuesta habiendo aplicado el filtrado bidireccional. Este tipo de filtrado anula por completo la respuesta de fase, por lo tanto se puede concluir que la distorsion observada en el primer filtrado era debido a ella.

## Filtrado con FIR

```python
ECG_f_win = sig.lfilter(fir_win, den_f, ecg_one_lead)

demora_fir = int((cant_coef - 1) / 2)

plot_ecg_filtered(ECG_f_win, 'Filtrado FIR', demora = demora_fir)
```

Se observa una buena respuesta del filtrado con FIR, sin la distorsion observada en el caso del IIR. Esto es debido a la fase lineal del filtro FIR, por lo que el filtrado bidireccional con este ultimo no es necesario para este caso.

De todas maneras, para fines demostrativos, se analizara por ultimo el caso del filtrado bidireccional con el FIR.

## Filtrado Bidireccional con FIR

```python
ECG_f_win_b = sig.filtfilt(fir_win, den_f, ecg_one_lead)

plot_ecg_filtered(ECG_f_win_b, 'Filtrado FIR Bidireccional')
```

Se observa practicamente la misma respuesta que con el filtrado normal con el FIR, por ende se concluye que este filtro no introduce distorsion de fase que deba ser corregida.

## Comparativa de Filtros

En esta ultima seccion, se presenta la señal original, y la filtrada con el IIR y el FIR, para observar diferencias entre ambos resultados.

```python
for ii in regs_interes:
    zoom_region = np.arange(np.max([0, ii[0]]), np.min([cant_muestras, ii[1]]), dtype='uint')
    
    plt.figure(figsize=(fig_sz_x, fig_sz_y), dpi= fig_dpi, facecolor='w', edgecolor='k')
    plt.plot(zoom_region, ecg_one_lead[zoom_region], label='ECG', linewidth=2)
    plt.plot(zoom_region, ECG_f_iir_b[zoom_region ], label='Filtrado IIR Bidireccional')
    plt.plot(zoom_region, ECG_f_win_b[zoom_region ], label='Filtrado FIR Bidireccional')
    
    plt.title('ECG filtering example from ' + str(ii[0]) + ' to ' + str(ii[1]) )
    plt.ylabel('Adimensional')
    plt.xlabel('Muestras (#)')
    
    axes_hdl = plt.gca()
    axes_hdl.legend()
    axes_hdl.set_yticks(())
            
    plt.show()
```

Se concluye que fue posible diseñar exitosamente dos filtros digitales con caracteristicas diferentes pero que respondieran a una misma plantilla: un IIR (infinite impulse response) y otro FIR (finite impulse response).

A la vez, se aplicaron estos filtros a una señal real de electrocardiograma de superficie, y se observó que ambos filtraron las bajas frecuencias exitosamente. En el caso del filtro IIR, se observo una fuerte distorsion de fase que debio ser corregida aplicando filtrado bidireccional, el cual anula la respuesta de fase.
